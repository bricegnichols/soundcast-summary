```python
import pandas as pd

%matplotlib inline
```

```python
df = pd.read_csv(r'J:\Projects\Soundcast\soundcast_dashboard\model_output\tours.csv')
df.columns
```

```python
# Save results by source as seperate df
df_dict = {}
for source in df.groupby('source').count().index.tolist():
    df_dict[source] = df[df['source'] == source]
```

# Tours by Purpose

```python
# Reformat to plot
dfplot = pd.DataFrame([df_dict[i].groupby('pdpurp').sum()['toexpfac'] for i in df_dict.keys()]).T
dfplot.columns = df_dict.keys()
print(dfplot.plot(kind='bar'))
```

# Tour Mode share

```python
dfplot = pd.DataFrame([df_dict[i].groupby('tmodetp').sum()['toexpfac'] for i in df_dict.keys()]).T
dfplot.columns = df_dict.keys()
print(dfplot.plot(kind='bar'))
```
